#### Вопрос 1
## Магазин карандашей
Витя пошёл в магазин за карандашами, каждый из которых стоит N рублей. Всего Витя готов потратить на карандаши M рублей. Укажите, сколько карандашей он сможет купить и сколько денег у него останется в итоге.

#### ФОРМАТ ВВОДА: 
Целое положительное число N — сколько стоит один карандаш.

Целое положительное число M — сколько денег готов потратить Витя на покупку.

#### ФОРМАТ ВЫВОДА: 
Целое число: сколько карандашей можно купить и, через пробел в той же строке, целое число: сколько денег останется после покупки.

**Для примера:**

| Ввод | Результат ||
| :- | :-||
| 19 | 5 5 |
| 100 |
| 72 | 0 50 |
| 50 |

```python
N, M = int(input()), int(input())
print( M // N, M % N)
```

#### Вопрос 2
## Время домашней работы

Усидчивый Матвей планирует заниматься ровно N часов. Помогите ему понять, во сколько он закончит заниматься сегодня, если на электронных часах сейчас A:B минут?

#### ФОРМАТ ВВОДА: 
Целое число N — сколько часов планирует заниматься Матвей, N >= 0

Строка вида ЧЧ:ММ, указывающая, который сейчас час.

Известно, что Матвей закончит заниматься в тот же день, когда начнёт.

####ФОРМАТ ВЫВОДА: 
Строка вида ЧЧ:ММ

**Подсказка:**

Если у вас есть строка a = '10:30', то можно "вытащить" первые 2 символа с помощью a[:2] , а последние 2 - с помощью a[3:]

| Ввод | Результат ||
| :- | :-||
| 2 | 14:30 |

#### Вопрос 4
## Кэшбэк
Банк предложил Арсению следующие условия по кэшбэку: 1% на все покупки, 3% на покупки в категории "продукты", 5% в категории "аптеки". Помогите Арсению определить по данным о его покупке, какой кэшбэк он получит. 

#### ФОРМАТ ВВОДА 
Строка — название категории покупки Далее целое положительное число — стоимость покупки

#### Вопрос 5
## Поход в музей
Билет в музей стоит 500 рублей, а для тех, кто предъявит студенческий билет — 250. Помогите Андрею понять, сможет ли он попасть в музей. 

#### ФОРМАТ ВВОДА 
Вводится целое положительное число — сумма денег, которая есть у Андрея с собой 
Далее с новой строки вводится "Да", если у Андрея есть при себе студенческий билет, и "Нет" в противном случае.

| Ввод | Результат ||
| :- | :-||

#### Вопрос 6
## Путевки на Олимпийские игры
Спортсмен Славик хочет завоевать путевки на Олимпийские игры для своей страны. Для этого он участвует в чемпионате мира и чемпионате Европы. Правила таковы: 

Чтобы гарантированно получить путевки на Олимпийские игры, нужно войти в топ-20 чемпионата мира **И** топ-16 чемпионата Европы (оба условия должны выполниться одновременно). При этом, если войти в топ-10 хотя бы одного из чемпионатов, страна получает сразу 2 путевки, а если войти в топ-2 хотя бы одного из чемпионатов, то 3.
Если условия из п. 1 не выполнены, но спортсмен вошел в топ-24 чемпионата мира **ИЛИ** топ-20 чемпионата Европы, он отправляется на дополнительные отборочные соревнования, где у него будет еще один шанс. 
В ином случае спортсмен не получает путевку на Олимпийские игры для своей страны. 
Напишите программу, которая определяет, завоевал ли Славик путевки на Олимпийские игры. 

#### ФОРМАТ ВВОДА: 
Два целых положительные числа, каждое с новой строки — места Славика на чемпионате мира и чемпионате Европы соответственно. 

#### ФОРМАТ ВЫВОДА 
Если Славик завоевал путевки, вывести "Ура! Путевки на Олимпийские игры завоеваны!" и далее с новой строки вывести количество путевок. 
Если Славик проходит на дополнительный отбор, вывести "Нужно принять участие в дополнительных отборочных соревнованиях". 
Иначе вывести "К сожалению, в этот раз не удалось завоевать путевки на Олимпийские игры".

**Для примера:**

| Ввод | Результат ||
| :- | :-||
| 15 | Ура! Путевки на Олимпийские игры завоеваны! |
| 9 | 2 |

```python
def check_win(world, euro):
    amount = 0
    is_choice = False

    if world <= 20 and euro <= 16:
        if world <= 2 or euro <= 2:
            amount = 3
        elif world <= 10 or euro <= 10:
            amount = 2
        else:
            amount = 1
    elif world <= 24 or euro <= 20:
        is_choice = True

    print_result(amount, is_choice)

def print_result(num, choice):
    if num:
        print("Ура! Путевки на Олимпийские игры завоеваны!", num, sep="\n")
    elif choice:
        print("Нужно принять участие в дополнительных отборочных соревнованиях")
    else:
        print("К сожалению, в этот раз не удалось завоевать путевки на Олимпийские игры")

top_w, top_e = int(input()), int(input())
check_win(top_w, top_e)
```

#### Вопрос 7
## Счастливые числа
Олег считает своими счастливыми числами все, которые делятся на 4 и на 6 одновременно (например, 12). Олег хочет написать программу, которая принимает на ввод число и выводит все счастливые числа Олега в промежутке от 1 до этого числа включительно. Если таких чисел нет, то программа ничего не выводит.

#### ФОРМАТ ВВОДА 
Целое положительное число N. 
#### ФОРМАТ ВЫВОДА 
Все числа в промежутке от 1 до N включительно, которые являются счастливыми для Олега, каждое с новой строки.

**Для примера:**

| Ввод | Результат ||
| :- | :-||
| 30 | 12 |
|| 24 |
| 100 | 12 |
|| 24 |
|| 36 |
|| 48 |
|| 60 |
|| 72 |
|| 84 |
|| 96 |

```python
N = int(input())
print(*[num for num in range(1, N + 1) if num % 4 == 0 and num % 6 == 0], sep="\n")
```

#### Вопрос 8
## Ежемесячные траты
Вася заметил, что с каждым месяцем у него все хуже получается откладывать деньги на новый ноутбук. Он подумал, что это может произойти по той причине, что с каждым месяцем он тратит все больше и больше. Проверьте, действительно ли в каждый следующий месяц Вася тратил больше, чем в предыдущий.

#### ФОРМАТ ВВОДА
Целые неотрицательные числа, не превосходящие 1 миллион, каждое с новой строки, затем отрицательное число, показывающее конец ввода. 
#### ФОРМАТ ВЫВОДА
True, если в каждый следующий месяц Вася тратил больше, чем в предыдущий (без учета последнего введенного числа, означающего конец ввода), и False в противном случае. 
Если были введены расходы всего за ноль или один месяц, ответ True.

**Для примера:**

| Ввод | Результат ||
| :- | :-||
| 20000 | True |
| 22000 |
| 28000 |
| -5 |

```python
flag, start = True, 0

while flag:
    amount = int(input())
    if amount < 0:
        break
    elif amount > start:
        start = amount
    else:
        flag = False

print(flag)
```

#### Вопрос 9
## Круглый ковер
Макс хочет купить в квадратный зал такой круглый ковер, чтобы он помещался точно от одной стены до другой. Но вот незадача: в магазине ковры продают по их периметру. Помогите Максу рассчитать периметр нужного ковра для зала, если ему известна площадь зала. 
Периметр круга считается по формуле pi * d, где d — диаметр круга, а pi — переменная из модуля math (в ней записано число Пи). Также для решения этой задачи понадобится функция sqrt из модуля math, которая извлекает квадратный корень из числа. 

#### ФОРМАТ ВВОДА 
Целое число — площадь зала. 
#### ФОРМАТ ВЫВОДА 
Вещественное число — периметр нужного ковра.

**Для примера:**

| Ввод | Результат ||
| :- | :-||
| 4 | 6.283185307179586 |

```python
from math import pi, sqrt

s_hall = int(input())
s_carpet = pi * sqrt(s_hall)
print(s_carpet)
```
