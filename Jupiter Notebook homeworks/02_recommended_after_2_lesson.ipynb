#### Вопрос 1
Кот Матроскин хочет отправить дяде Федору и Шарику новогодние телеграммы. Для этого он может воспользоваться двумя тарифами: обычный (2 рубля за символ) или премиум (3 рубля за символ). 

Кроме того, за телеграммы длиннее 50 символов полагается доплата: +10 рублей в обычном тарифе и +15 в премиум. Помогите Матроскину рассчитать стоимость отправки телеграммы. 

#### ФОРМАТ ВВОДА 
Текст телеграммы в одну строку. 
Далее с новой строки вводится тариф отправки — строка **"обычный"** или **"премиум"**. 

#### ФОРМАТ ВЫВОДА 
Целое число — стоимость отправки телеграммы в рублях.

#### ПОДСКАЗКА: можно использовать функцию len() для поиска длины строки.
Например, len('привет') выведет 6.

**Для примера:**

| Ввод | Результат ||
| :- | :-||
| ПОЗДРАВЛЯЮ ТЕБЯ ШАРИК ТЫ БАЛБЕС | 62 ||
| обычный ||

```python
msg, rate = input(), input()
count = len(msg)
if rate == "обычный":
    if len(msg) <= 50:
        print(count * 2)
    else:
        print(count * 2 + 10)
elif rate == "премиум":
    if len(msg) <= 50:
        print(count * 3)
    else:
        print(count * 3 + 15)
```

#### Вопрос 2
Алиса в течение года проверяет, какая температура в Сочи. Она хочет прилететь туда, когда температура там будет либо от +20 до +30 градусов Цельсия включительно, чтобы искупаться в море, либо когда температура будет от -5 до +5 градусов Цельсия включительно, чтобы ездить в горы и кататься на лыжах. Помогите Алисе определить, можно ли лететь в Сочи. 

#### ФОРМАТ ВВОДА 

Целое число — количество градусов по Цельсию в Сочи
#### ФОРМАТ ВЫВОДА 

Если Алиса может лететь в Сочи, выводится строка "Можно лететь!". 
Иначе выводится строка "Не время для отпуска".

***Для примера:***

| Ввод | Результат ||
| :- | :-||
| 25 | Можно лететь! ||

```python
degree = int(input())

if 20 <= degree <= 30 or -5 <= degree <= 5:
    print("Можно лететь!")
else:
    print( "Не время для отпуска")
```

#### Вопрос 3
## Банкомат
Банкомат выдает суммы, только кратные 100 рублям, потому что купюр мельче у него нет. Выведите на экран "ДА", если банкомат сможет выдать введенную сумму, и "НЕТ", если не сможет. 

#### ФОРМАТ ВВОДА
Вводится требуемая сумма для вывода средств, целое число без копеек. 
#### ФОРМАТ ВЫВОДА
Строка "ДА" или "НЕТ", в зависимости от того, сможет ли банкомат выполнить операцию.

**Для примера:**

| Ввод | Результат ||
| :- | :-||
| 1365 | НЕТ |

```python
money = int(input())

if not money % 100:
    print("ДА")
else:
    print("НЕТ")
```

#### Вопрос 4
## Какой это век?
Определите по номеру года, в каком веке этот год. Не забывайте, что, например, 21 век начался с 2001 года, а 2000 год относится к 20 веку. 

#### ФОРМАТ ВВОДА 
Целое число от 1 до 2100 — год 
#### ФОРМАТ ВЫВОДА 
Целое число — век, к которому относится введенный год

**Для примера:**

| Ввод | Результат ||
| :- | :-||
| 2020 | 21 |

```python
year = int(input())

if year % 100:
    print(year // 100 + 1)
else:
    print(year // 100)
```
