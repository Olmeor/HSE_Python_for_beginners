#### Вопрос 1
## Иероглифы
Вася учит японские слова и делает карточки. Он вводит слова парами: слово на русском языке и японский иероглиф. Нужно считать пары в словарь, где ключом будет русское слово, а иероглиф — значением. Когда пары закончатся, Вася введет "все". Затем Вася проверяет, все ли слова он записал на карточки. Ваша программа должна проверять в словаре наличие иероглифа для введенного русского слова и либо выводить этот иероглиф, либо сообщение "Такого слова нет".

#### ФОРМАТ ВВОДА
Строки с парами: русское слово и иероглиф
Пары слов вводятся до тех пор, пока не введена строка "все"
Затем вводится строка с русским словом, перевод которого нужно найти

#### ФОРМАТ ВЫВОДА
Строка с иероглифом (перевод слова) или строка "Такого слова нет"

```
ЧТО ДЕЛАЕМ:

Реализовать цикл while — пока строка из п. 1 не равна строке "все":
превратить строку из п. 1 в список,
добавить в словарь новый элемент, где ключ — это первый элемент списка, а значение — второй,
считать новую строку с парой слов.
После завершения работы цикла, считать с клавиатуры в новую переменную слово, которое хотим перевести.
Проверить условие: если это слово есть в словаре из п. 2, то вывести его перевод. В ином случае напечатать "Такого слова нет".
```
</details>

**Для примера:**

| Ввод | Результат |
| :- | :-|
| молоко 牛乳 | 日 |
| солнце 日 |
| конституция 憲法 |
| имитация 模倣 |
| все |
| солнце |

````python
dct = {}
while True:
    s = input()
    if s == 'все':
        break
    else:
        k, v = s.split()
        dct[k] = v

print(dct.get(input(), "Такого слова нет"))
````

Для примера:
