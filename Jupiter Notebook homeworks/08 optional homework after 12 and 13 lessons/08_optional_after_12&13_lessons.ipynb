````python
dct_name = {"капучино": 60, "латте": 60, "раф": 80}
dct_size = {"маленький": 1, "средний": 2, "большой": 3}
total = 0

with open('coffee-statistics.csv', 'r', encoding='utf-8') as file_in:
    for row in file_in:
        name, size = row.strip().split(';')
        total += dct_name[name] * dct_size[size]

print(total)
````
