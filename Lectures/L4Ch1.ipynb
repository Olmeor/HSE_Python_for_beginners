## Срезы строк

Иногда нам нужно обратиться к последовательности букв. Такая возможность называется **срезом строки**. Срезы похожи на обращение к символу строки по его индексу, только теперь будет не один индекс в квадратных скобках, а два (или три) — индекс символа, с которого нужно начать срез, перед символом с каким индексом будет нужно остановиться и – опционально – с каким шагом выбирать символы.

<img src='https://www.bestprog.net/wp-content/uploads/2019/12/11_01_02_01_06_03_01_.jpg'>

Синтаксис среза в общем виде выглядит так:
`s[START:STOP:STEP]`

* START — индекс первого элемента в срезе
* STOP — индекс элемента списка, перед которым срез должен закончиться. Сам элемент с индексом STOP не включается в срез
* STEP — шаг, с которым нужно выбрать элементы для среза

**Несколько вариантов:**

* `[:]/[::]` — весь список
* `[::2]` — нечетные по порядку элементы
* `[1::2]` — четные по порядку элементы
* `[::-1]` — все элементы в обратном порядке
* `[4:]` — все элементы, начиная с пятого по счету, с четвертого по индексу
* `[:4]` — все элементы, не доходя до пятого по счету, до четвертого по индексу
* `[-2:1:-1]` — все элементы от предпоследнего до третьего в обратном порядке

```python
s = 'abcdef'
```

```python
s[::]
```

```python
s[::2]
```

```python
s[1::2]
```

```python
s[:4]
```

```python
print('abcdef'[0:2]) # напечатай символы, начиная с первого символа в строке "abcdef", и остановись перед третьим
print('abcdef'[1:3]) # напечатай символы, начиная со второй буквы в слове "abcdef", и остановись перед четвёртым
```

Если мы хотим сделать срез "от начала строки" или "до конца строки", мы можем не писать соответствующие индексы:

```python
print('abcdef'[:3]) # напечатай символы, начиная с начала строки "abcdef" и остановись перед четвёртой буквой

print('abcdef'[3:]) # напечатай символы, начиная с четвёртого символа строки "abcdef"

print('abcdef'[:])  # наконец, мы можем вообще не указать границы среза — тогда 
                   # компьютер попробует угадать сам, что нам нужно вывести;
                   # например, сейчас он выведет всю строку от начала и до конца
```

Кроме того, что мы можем обратиться к части строки, мы можем ещё и обратиться к символам не подряд. Например, выведем каждую вторую букву алфавита:

```python
alphabet = 'АБВГДЕЁЖЗИЙКЛМНОПРСТУФХЦЧШЩЪЫЬЭЮЯ'
print(alphabet[0:33:2]) # начнём с первой буквы и остановимся на 33-й, 
                        # будем брать каждую вторую букву
```

Благодаря последнему аргументу мы можем управлять порядком чтения букв в строке. Например, можем попросить компьютер вывести алфавит в обратном порядке:

```python
print(alphabet[::-1])
```

Можем попросить вывести каждую вторую букву из первых 10 букв алфавита в обратном порядке:

```python
print(alphabet[9::-2]) # начинаем с 10-й буквы (компьютер всё считает с 0), заканчиваем на 0-й букве,
                       # берём каждую вторую букву, идя назад
```

### Пример задачи со срезом строк

Дана строка, нужно проверить является ли она палиндромом. Строка является палиндромом, если одинаково читается в обоих
направлениях (слева направо и справа налево). Для простоты будем считать, что знаков препинания в строке нет.

Пример палиндрома: 'а роза упала на лапу азора'

```python
check = input('Введите палиндром для проверки: ')

data = check.replace(' ', '').lower()

if data == data[::-1]:
    print('Это палиндром')
else:
    print('Это не палиндром')
```

## Цикл While

**While** - вид цикла в Python

Циклы нужны в тех случаях, когда нужно много раз сделать однотипную задачу. Например, посчитать одни и те же показатели для отчета для большого количества одинаковых табличек. Это экономит время и позволяет автоматизировать рутинную работу (похоже чем-то на конвейер).

Цикл **while** (в переводе - **пока**) позволяет выполнить одну и ту же последовательность действий, **пока** проверяемое условие истинно.

<img src="https://i.ibb.co/w0dP6n0/Untitled-Frame-1-15.jpg">

Немного более формальная и серьезная схема:

<img src="https://upload.wikimedia.org/wikipedia/commons/thumb/4/43/While-loop-diagram.svg/1200px-While-loop-diagram.svg.png" width="50%" height="30%">

**Зачем**:
<br>
Как правило, цикл while используется, когда мы знаем, при каких случаях можно повторить наш код.

**Условие**:
<br>
Условие записывается до тела цикла и проверяется до выполнения тела цикла. <br>

**Еще раз подробно:**<br>
<br>
При выполнении цикла while сначала проверяется условие. Если оно ложно, то выполнение цикла прекращается и управление передается на следующую инструкцию после тела цикла while. Если условие истинно, то выполняется инструкция, после чего условие проверяется снова и снова выполняется инструкция. Так продолжается до тех пор, пока условие будет истинно. Как только условие станет ложно, работа цикла завершится и управление передастся следующей инструкции после цикла.

Синтаксис цикла **while** в простейшем случае выглядит так:

`while <<условие>> :
    <<блок инструкций>>`

Пример использования `while` в [игре](https://reeborg.ca/reeborg.html?lang=en&mode=python&menu=worlds%2Fmenus%2Freeborg_intro_en.json&name=Hurdle%202&url=worlds%2Ftutorial_en%2Fhurdle2.json).

### Задача № 3
Используя цикл while, выведите на экран для числа 2 его степени от 0 до 10.

```python
n = 0

while n <= 10:
    print('Число 2 в степени',n, '=', 2**n)
    n+=1
```

### Задача № 5
Дано целое число, не меньшее 2. Выведите его наименьший натуральный делитель, отличный от 1.

```python
n = int(input())

i = 2

while n%i != 0:
    i += 1
    print(i)
    
print('Итог', i)
```

### Задача № 6

А теперь напишем маленькую игру-угадайку. Программа будет загадывать целое число от 1 до 100, а пользователь его угадывать. Как программа будет загадывать число? Выбирать случайным образом из интервала [1, 100] (на самом деле псевдослучайным образом, так как абсолютной случайности не получится, генерирование чисел происходит по фиксированным алгоритмам).

```python
from random import randrange # импортируем функию randrange
```

```python
n = randrange(1, 101) # n и есть загаданное число
```

Осталось написать цикл. До тех пор, пока пользователь не угадает число, программа не будет останавливаться, но зато она будет давать подсказки: если введенное пользователем число больше загаданного, то будет выводиться сообщение "Вы ввели слишком большое число.", если меньше – "Вы ввели слишком маленькое число."

```python
while True:
    
    guess = int(input("Ваша попытка: "))
    
    if guess == n:
        print("Вы выиграли!")
        break
    elif guess > n:
        print("Вы ввели слишком большое число.")
    else: 
        print("Вы ввели слишком маленькое число.")
```

В коде выше в `while` мы не написали никакого условия явно, вместо этого мы написали `while True`. Это выражение означает «до тех пор, пока мы не вышли из цикла». В нашем случае это равносильно «до тех пор, пока не столкнулись с `break`», то есть пока наш ответ не совпал с загаданным числом.
